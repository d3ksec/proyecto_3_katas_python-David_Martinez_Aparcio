1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.


```python
texto = "Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum."

def contar_frecuencia_letras(texto):
    frecuencias = {}
    for caracter in texto:
        if caracter != " ":
            if caracter in frecuencias:
                frecuencias[caracter] += 1
            else:
                frecuencias[caracter] = 1
    return frecuencias

resultado = contar_frecuencia_letras(texto)

print(resultado)
```

4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función map().

```python
lista_1 = [10, 20, 30]
lista_2 = [2, 5, 10]

def diferencia_listas(listaa, listab):
    diferencia = list(map(lambda x, y: x - y, listaa, listab))
    return diferencia

resultado_diferencia_listas = diferencia_listas(lista_1, lista_2)

print(resultado_diferencia_listas)
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5). La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def nota_media(notas, nota_aprobado=5):
    media = sum(notas) / len(notas)
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"

    return(media, estado)

notas_redes = [6.5, 4.0, 8.2, 5.5, 7.0]

notas1 = nota_media(notas_redes)
print(notas1)
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.


```python
def calcular_factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * calcular_factorial(n - 1)

resultado = calcular_factorial(4)
print(resultado)
```

8. Escribe un programa que pida al usuario dos números e intente dividirlos. Si el usuario ingresa un valor no numérico o intenta dividir por cero, maneja esas excepciones de manera adecuada y muestra un mensaje indicando si la división fue exitosa o no.

```python
def division(numa, numb):
    try:
        numa = float(numa)
        numb = float(numb)
        resultado = numa / numb
        return f"División exitosa: {resultado}"
    except ValueError:
        return "Error: Ingrese un dato numérico válido."
    except ZeroDivisionError:
        return "Error: No se puede dividir entre cero."

numero1 = input("Introduce el primer número: ")
numero2 = input("Introduce el segundo número: ")

print(division(numero1, numero2))
```

9. Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]. Usa la función filter().


```python
animales_prohibidos = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
animales = ["Perro", "Mapache", "Gato", "Loro", "Tigre", "Hámster", "Cocodrilo", "Conejo", "Serpiente Pitón"]

def detector_mascotas(animales, animales_prohibidos):
    mascotas = list(filter(lambda animal: animal not in animales_prohibidos, animales))
    return mascotas


resultado = detector_mascotas(animales, animales_prohibidos)
print(resultado)
```

```python
numeros = [42, 15, 88, 7, 34, 59, 21, 93, 10, 65]

def promedio(numeros):
    if len(numeros) == 0:
        return ("Error: la lista está vacía")
    else:
        return (sum(numeros) / len(numeros))


print(promedio(numeros))

print(promedio([]))
```

13. Genera una función que, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas. Usa la función map().

```python
def caracteres_tupla(texto):
    caracteres_unicos = set(texto)
    resultado = list(map(lambda letras: (letras.upper(), letras.lower()), caracteres_unicos))
    return resultado

letras = input("Introduce caracteres")
tupla = caracteres_tupla(letras)

print(tupla)
```

14. Crea una función que retorne las palabras de una lista que comiencen con una letra en específico. Usa la función filter().

```python
def filtro_palabras(lista_palabras, letra_inicial):
    letra_inicial = letra_inicial.lower() #Para evitar problemas de mayusculas y minusculas

    resultado = list(filter(lambda palabra: palabra.lower().startswith(letra_inicial), lista_palabras))    

    return resultado

set_up = ["Teclado", "Monitor", "Mesa", "Ratón", "Micrófono", "Router"]

palabras_filtradas = filtro_palabras(set_up, input("Introduce una letra")) #Por ejemplo la m
print(palabras_filtradas)
```

16. Escribe una función que tome una cadena de texto y un número entero n como parámetros y devuelva una lista de todas las palabras que sean más largas que n. Usa la función filter().

```python
def palabras_mas_largas(texto, n):

    palabras = texto.split()

    resultado = list(filter(lambda palabra: len(palabra) > n, palabras))

    return resultado

frase = "No. Yo soy tu padre"
longitud_min = int(input("Introduce una longitud mínima: "))

palabras_largas = palabras_mas_largas(frase, longitud_min)
print(palabras_largas)
```

17. Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2] corresponde al número 572. Usa la función reduce().

```python
from functools import reduce

def lista_a_numero(lista_digitos):
    resultado = reduce(lambda acumulador, digito: acumulador * 10 + digito, lista_digitos)
    return resultado

digitos = [5,7,2]
numero_final = lista_a_numero(digitos)

print(numero_final)
```
